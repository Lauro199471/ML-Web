```python
# imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
# generate random data-set
np.random.seed(0)
x = np.random.rand(100, 1)
y = 2 + 3 * x + np.random.rand(100, 1)
```

```python
def linearRegression(X,Y):
    # Add ones col to feature matrix to get y-intercept
    n,m = X.shape
    oneCol = np.ones((n,1))
    X_with_Ones = np.hstack((oneCol,X))
    
    # find x dagger 
    x_dagger = np.dot((np.linalg.inv(np.dot(X_with_Ones.T,X_with_Ones))), X_with_Ones.T)
    w = np.dot(x_dagger,Y)
    
    return w[0,0],w[1,0]

y_intercept,slope = linearRegression(x,y)
print("y_intercept: ", y_intercept)
print("slope:" , slope)
```

```python
# plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(x,y,color='blue',edgecolors='black', label='Scatter Plot')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
ax.set(xlabel='x', ylabel='y')
ax.set_facecolor('#f0f0f0')
```

```python
slope = 2.936551060203767
y_intercept = np.linspace(0,5,10000)
#calculuate sum of squared residual
sum_squaredList = []
for i in range(len(y_intercept)):
    squared_residualList = []
    for j in range(len(x)):
        predication = (slope*x[j]) + y_intercept[i]
        residual = y[j] - predication
        squared_residual = np.power(residual,2)
        squared_residualList.append(squared_residual)
    sum_squaredList.append(np.sum(squared_residualList))
```

```python
# plot
fig, ax = plt.subplots(figsize=(10, 5))
ax.scatter(y_intercept,sum_squaredList,color='blue', label='Sum of Squared Residuals',s=0.5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(color='grey', linestyle='-', linewidth=0.1, alpha=0.5)
ax.set(xlabel='Intercept', ylabel='Sum of Squared Residuals')
plt.axvline(x=2.5580800241570225,linestyle='--',label="best y-intercept")
ax.set_facecolor('#f0f0f0')
ax.legend();
```
