Logistic Regression is used for binary classification problems. Unlike it’s cousin Linear Regression which outputs continuous values, Logistic regression outputs probabilities within range 0–1 which represent the liklihood a data point is either A or B, True or False, Pass or Fail, etc

First we will load our libraries and dataset which will be the classic Iris classification problem.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Plot Function
This will just plot our data

```python
def plot(X,Y,slope=0,y_intercept=0):
    # If no slope and y_intercept is passed then plot scatter only
    if(slope == 0.0 and y_intercept == 0.0):
        fig, ax = plt.subplots(figsize=(10, 5))
        # Plot Scatter points
        ax.scatter(X[:,0][Y[:,0] == 1], X[:,1][Y[:,0] == 1] , color='blue', label="Passed")
        ax.scatter(X[:,0][Y[:,0] == 0], X[:,1][Y[:,0] == 0] , color='red' , label="Not Passed")
        
        # For better graph
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
        ax.set(xlabel='Exam 1 Score', ylabel='Exam 2 Score', title='Acceptance')
        ax.set_facecolor('#f0f0f0')
        ax.legend()
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    else:
        fig, ax = plt.subplots(figsize=(10, 5))
        max_x = np.max(X)
        min_x = np.min(X)
        
        x = np.linspace(min_x,max_x,(max_x*100))
        y = (slope*x) + y_intercept
        
        #Plot line
        ax.plot(x,y,color='green',label = 'Regression Line')
        # Plot Scatter points
        ax.scatter(X[:,0][Y[:,0] == 1], X[:,1][Y[:,0] == 1] , color='blue', label="Passed")
        ax.scatter(X[:,0][Y[:,0] == 0], X[:,1][Y[:,0] == 0] , color='red' , label="Not Passed")
        # Plot Means
        plt.axvline(x=np.mean(X[:,0]),linestyle='--',label="X mean",color='purple')
        plt.axhline(y=np.mean(X[:,1]),linestyle='--',label="Y mean")
        
        # For better graph
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.5)
        ax.set(ylim=(20, 110),xlabel='Exam 1 Score', ylabel='Exam 2 Score', title='Acceptance')
        ax.set_facecolor('#f0f0f0')
        ax.legend()
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

### Load Data Set
The data consists of marks of two exams for 100 applicants. The target value takes on binary values 1,0. <br>1 means the applicant was admitted to the university whereas 0 means the applicant didn't get an admission.<br> The objective is to build a classifier that can predict whether an application will be admitted to the university or not.

```python
data = pd.read_csv('admission.csv')

print('Head():\n',data.head(),'\n')
print('Tail():\n',data.tail(),'\n')
print('Data Shape: ',data.shape)
```

## Seperate Data

```python
# X = feature values, all the columns except the last column
X = np.array([data['Exam 1 Score'].values,data['Exam 2 Score'].values]).T
Y = np.array([data['Acceptance'].values]).T
print(X.shape,Y.shape)
```

```python
plot(X,Y)
```

```python
def linearRegression(X,Y):
    # Add ones col to feature matrix to get y-intercept
    m,n = X.shape
    print(n)
    oneCol = np.ones((n,1))
    X_with_Ones = np.hstack((oneCol,X))
    
    # find x dagger 
    x_dagger = np.dot((np.linalg.inv(np.dot(X_with_Ones.T,X_with_Ones))), X_with_Ones.T)
    w = np.dot(x_dagger,Y)
    print(w)
    return 0,1
```

```python
y_intercept,slope = linearRegression(np.array([X[:,0]]),np.array([X[:,1]]))
print("y_intercept: ", y_intercept)
print("slope:" , slope)
```

```python
plot(X,Y,slope,y_intercept)
```
